4. HISTOGRAM FOR EACH FEATURE

```python
import matplotlib.pyplot as plt

# Create histograms for each feature
df.drop("Species", axis=1).hist(bins=15, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

5. BOXPLOT FOR EACH FEATURE

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Create boxplots for each feature with respect to species
plt.figure(figsize=(12, 8))

# SepalLengthCm
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Boxplot of Sepal Length')

# SepalWidthCm
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Boxplot of Sepal Width')

# PetalLengthCm
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Boxplot of Petal Length')

# PetalWidthCm
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Boxplot of Petal Width')

plt.tight_layout()
plt.show()
```

To compare the distributions and identify outliers in the Iris dataset, we can analyze the boxplots and histograms generated previously. Below is an explanation of how to interpret the boxplots and identify any potential outliers.

Boxplot Analysis and Outlier Identification :

1. Boxplot Interpretation:
   - The box: The middle 50% of the data is represented by the box, where the bottom of the box is the 25th percentile (Q1) and the top of the box is the 75th percentile (Q3).
   - The line inside the box (median): This represents the 50th percentile of the data.
   - Whiskers:The lines extending from the box. These represent data points within 1.5times the interquartile range(IQR) from the Q1 & Q3.
   - Outliers: Data points outside the whiskers (i.e., beyond 1.5 times the IQR) are considered outliers and are typically marked as individual points.

2. Outliers in each feature can be identified by:
   - Checking if there are any points outside the whiskers in the boxplot.
   - Looking at the extreme values in the histograms.
