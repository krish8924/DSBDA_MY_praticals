1. IMPORT ALL REQUIRED PYTHON LIBRARIES

```python
import numpy as np
import pandas as pd
```

2. LOAD THE DATASET INTO PANDAS DATA FRAME

```python
df = pd.read_csv('Iris.csv')
df.head()
```

3. Data Preprocessing : CHECK FOR MISSING VALUES IN THE DATA FRAME

```python
df.isnull().sum()
```

4. DISPLAY INITIAL STATISTICS, VARIABLE DESCRIPTION AND TYPES, DIMENSION OF THE DATAFRAME

```python
print("Initial Statistics :")
print(df.describe())
print("\n Variable description and types : ")
print(df.dtypes)
print("\n Dimension of dataframe :")
print(df.shape)
```

5. DATA FORMATING AND DATA NORMALIZATION

```python
print("Data formating :")
print(df.dtypes)
# Example: Convert 'SepalLengthCm' to float (already is in most cases)
df['SepalLengthCm'] = df['SepalLengthCm'].astype(float)
print(df.dtypes)
print("\n Data Normalization :")
# normalize numeric columns
cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for col in cols:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

df.head()
```

6. TURN CATEGORICAL VARIABLES INTO QUANTITATIVE VARIABLES 

```python
# Convert categorical column (Species) into numeric labels
df['Species_Label'] = df['Species'].astype('category').cat.codes

# View result
df[['Species', 'Species_Label']].head()
```
