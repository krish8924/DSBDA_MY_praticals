Create an “Academic performance” dataset of students and perform the following operations using Python :

1. IMPORT ALL THE REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
```

2. LOAD THE DATASET USING PANDAS

```python
df = pd.read_csv('student_score2.csv')
df.head()
```

3. CHECK FOR INCONSISTENCIES

```python
df.isnull().sum()
```

4. HANDLING INCONSISTENCIES

```python
df.fillna({"reading score": df["reading score"].mean()}, inplace=True)
print(df.isnull().sum())
```

5. DETECTING OUTLIERS USING IQR

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Find outliers
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
outliers.sum()
```

6. REMOVE OUTLIERS

```python
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers.shape
```

7. DATA TRANSFORMATION : LOG TRANSFORMATION

```python
# Apply log transformation on numeric columns
for col in df_no_outliers.select_dtypes(include=np.number):
    df_no_outliers[col] = np.log1p(df_no_outliers[col])

df_no_outliers.head()
```
